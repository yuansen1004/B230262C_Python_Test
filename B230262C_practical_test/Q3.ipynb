```python
import pandas as pd
import matplotlib.pyplot as plt

coffee_df = pd.read_csv('coffee.csv')

total_orders = coffee_df.groupby(['product_category', 'product_type'])['transaction_qty'].sum().reset_index()

total_orders.columns = ['category', 'product', 'total_order']

print(total_orders)

plt.figure(figsize=(10, 6))
for category in total_orders['category'].unique():
    subset = total_orders[total_orders['category'] == category]
    plt.bar(subset['product'], subset['total_order'], label=category)

plt.xlabel('Product')
plt.ylabel('Total Orders')
plt.title('Total Orders by Product Type')
plt.xticks(rotation=45, ha='right')
plt.legend()
plt.show()
```
